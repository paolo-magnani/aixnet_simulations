```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import ast

df = pd.read_csv('final_results.csv')
display(df)

for col in ["light", "heavy", "frac_l", "frac_h"]:  # Adjust column names if needed
    df[col] = df[col].apply(lambda x: ast.literal_eval(x)[0] if isinstance(x, str) else x)
```

```python
colors ={'8': 'red', '16': 'green', '32': 'blue'}
energ = [8, 16, 32, 200]
plt.figure(figsize=(10,6))
for i in range(len(energ[:-1])):
    ene = energ[i]
    plt.hist(df['frac_l'][df['emin']==ene], histtype='step', stacked=False, fill=False, color=colors[f'{ene}'], lw=4, alpha=0.9)
    plt.axvline(np.mean(df['frac_l'][df['emin']==ene]), ls='--', lw=4, label=f'$[{ene}, {energ[i+1]}] \, \mathrm{{EeV}}: {np.mean(df['frac_l'][df['emin']==ene]):.2f}$', color=colors[f'{ene}'])
    print(np.mean(df['frac_l'][df['emin']==ene]))
plt.legend()
plt.xlabel('fraction of lightest events')
plt.ylabel('# simulated data sets')
plt.show()


plt.figure(figsize=(10,6))
for i in range(len(energ[:-1])):
    ene = energ[i]
    plt.hist(df['frac_h'][df['emin']==ene], histtype='step', stacked=False, fill=False, color=colors[f'{ene}'], lw=4, alpha=0.9)
    plt.axvline(np.mean(df['frac_h'][df['emin']==ene]), ls='--', lw=4, label=f'$[{ene}, {energ[i+1]}] \, \mathrm{{EeV}}: {np.mean(df['frac_h'][df['emin']==ene]):.2f}$', color=colors[f'{ene}'])
    print(np.mean(df['frac_h'][df['emin']==ene]))
plt.legend()
plt.xlabel('fraction of heaviest events')
plt.ylabel('# simulated data sets')
plt.show()
```
