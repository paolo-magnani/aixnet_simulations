```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import ast

df = pd.read_csv('partial_results_univ.csv')
display(df)

# for col in ["light", "heavy", "frac_l", "frac_h"]:  # Adjust column names if needed
#     df[col] = df[col].apply(lambda x: ast.literal_eval(x)[0] if isinstance(x, str) else x)
```

```python
colors ={'8': 'red', '16': 'green', '32': 'blue'}
energ = [8, 16, 32, 200]
plt.figure(figsize=(10,6))
for i in range(len(energ[:-1])):
    ene = energ[i]
    plt.hist(df['frac_l'][(df['emin']==ene)&(df['frac_l']!=0)], histtype='step', stacked=True, fill=False, color=colors[f'{ene}'], lw=4, alpha=0.8)
    plt.axvline(np.mean(df['frac_l'][(df['emin']==ene)&(df['frac_l']!=0)]), ls='--', lw=4, label=f'$[{ene}, {energ[i+1]}] \, \mathrm{{EeV}}: {np.mean(df['frac_l'][df['emin']==ene]):.2f}$', color=colors[f'{ene}'])
    print(np.mean(df['frac_l'][(df['emin']==ene)&(df['frac_l']!=0)]))
plt.legend()
plt.xlim(0.01, 0.8)
plt.xlabel('fraction of lightest events')
plt.ylabel('# simulated data sets')
plt.show()


plt.figure(figsize=(10,6))
for i in range(len(energ[:-1])):
    ene = energ[i]
    plt.hist(df['frac_h'][(df['emin']==ene)&(df['frac_h']!=0)], histtype='step', fill=False, color=colors[f'{ene}'], lw=4, alpha=0.8)
    plt.axvline(np.mean(df['frac_h'][(df['emin']==ene)&(df['frac_h']!=0)]), ls='--', lw=4, label=f'$[{ene}, {energ[i+1]}] \, \mathrm{{EeV}}: {np.mean(df['frac_h'][df['emin']==ene]):.2f}$', color=colors[f'{ene}'])
    print(np.mean(df['frac_h'][(df['emin']==ene)&(df['frac_h']!=0)]))
plt.legend()
plt.xlim(0.01, 0.8)
plt.xlabel('fraction of heaviest events')
plt.ylabel('# simulated data sets')
plt.show()


# da ChatGPT perchÃ© sono pigro cazzo
fig, axs = plt.subplots(1, 2, figsize=(18, 6), sharey=True)

# First plot: fraction of lightest events
for i in range(len(energ[:-1])):
    ene = energ[i]
    data_l = df['frac_l'][(df['emin'] == ene) & (df['frac_l'] != 0)]
    mean_l = np.mean(data_l)
    axs[0].hist(data_l, histtype='step', stacked=True, fill=False,
                color=colors[f'{ene}'], lw=4, alpha=0.8)
    axs[0].axvline(mean_l, ls='--', lw=4, color=colors[f'{ene}'],
                   label=f'$[{ene}, {energ[i+1]}] \, \mathrm{{EeV}}: {mean_l:.2f}$')
    print(mean_l)

axs[0].legend()
axs[0].set_xlim(0.01, 0.8)
axs[0].set_xlabel('fraction of lightest events')
axs[0].set_ylabel('# simulated data sets')

# Second plot: fraction of heaviest events
for i in range(len(energ[:-1])):
    ene = energ[i]
    data_h = df['frac_h'][(df['emin'] == ene) & (df['frac_h'] != 0)]
    mean_h = np.mean(data_h)
    axs[1].hist(data_h, histtype='step', fill=False,
                color=colors[f'{ene}'], lw=4, alpha=0.8)
    axs[1].axvline(mean_h, ls='--', lw=4, color=colors[f'{ene}'],
                   label=f'$[{ene}, {energ[i+1]}] \, \mathrm{{EeV}}: {mean_h:.2f}$')
    print(mean_h)

axs[1].legend()
axs[1].set_xlim(0.01, 0.8)
axs[1].set_xlabel('fraction of heaviest events')
axs[1].set_ylabel('# simulated data sets')
axs[1].tick_params(labelleft=True)

# plt.tight_layout()
plt.show()


print(len(df['frac_l'][(df['frac_l']!=0)]))
```
